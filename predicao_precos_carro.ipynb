## Treinamento do modelo

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb
import sklearn.preprocessing
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
import time
```

```python
# Antes de treinar o modelo, vou utilizar a codificação de rótulos para as características 
le = LabelEncoder()
df_scaled = df.apply(le.fit_transform)
```

```python
# Agora vou escalar os dados
feature_names = ['VehicleType', 'RegistrationYear', 'Gearbox', 'Power', 'Model', 'Mileage', 'FuelType', 'Brand', 'NotRepaired', 'RegistrationMonth']

transformer_mas = sklearn.preprocessing.MaxAbsScaler().fit(df_scaled[feature_names].to_numpy())

df_scaled = df_scaled.copy()
df_scaled.loc[:, feature_names] = transformer_mas.transform(df_scaled[feature_names].to_numpy())
df_scaled
```

### Divididndo os dados em treino e teste

```python
# para o df
X = df.drop(['Price', 'DateCrawled', 'DateCreated', 'PostalCode', 'LastSeen'], axis = 1).to_numpy()
y = df['Price'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12345)
```

```python
# para o df_scaled
X_scaled = df_scaled.drop(['Price', 'DateCrawled', 'DateCreated', 'PostalCode', 'LastSeen'], axis = 1).to_numpy()
y_scaled = df_scaled['Price'].to_numpy()
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X_scaled, y_scaled, test_size=0.25, random_state=12345)
```

### Modelo LightGBM

```python
from sklearn.model_selection import GridSearchCV
train_data_scaled = lgb.Dataset(X_train_scaled, label=y_train_scaled)
test_data_scaled = lgb.Dataset(X_test_scaled, label=y_test_scaled)
```

```python
params = {
    'num_leaves': [31,100] ,
    'learning_rate' : [0.01, 0.05],
}
```

```python
model_hip = lgb.LGBMRegressor()

# Realiza a busca em grade dos melhores hiperparâmetros
grid_search = GridSearchCV(model_hip, param_grid=params, cv=5)
grid_search.fit(X_train_scaled, y_train_scaled)

# Imprime os melhores hiperparâmetros encontrados
print(grid_search.best_params_)
```

```python
params_model = {
    'metric': 'rmse',
    'num_leaves' : 31,
    'learning_rate' : 0.05,
}
```

```python
start_time = time.time()
model = lgb.train(params_model, train_data_scaled, valid_sets=[test_data_scaled], num_boost_round=1000, early_stopping_rounds=50)
end_time = time.time()
total_time_lbg = end_time - start_time
```

```python
y_pred_gbm = model.predict(X_test_scaled)
```

```python
rmse_gbm = mean_squared_error(y_test_scaled, y_pred_gbm, squared=False)
rmse_gbm
```

### Modelo Floresta Aleatória

```python
# Calibrando os hiperparâmetros
rf_model = RandomForestRegressor()

param_grid = {
    'n_estimators': [5, 10, 15], 
    'max_depth': [1, 5, 10]
}
```

```python
rf_grid = GridSearchCV(rf_model, param_grid, cv=5)
rf_grid.fit(X_train_scaled, y_train_scaled)
```

```python
print('Melhores hiperparâmetros: ', rf_grid.best_params_)
print('Melhor pontuação: ', rf_grid.best_score_)
```

```python
# Treinando o modelo
rf = RandomForestRegressor(max_depth = None, n_estimators=20, random_state=12345)
start_time = time.time()

rf.fit(X_train_scaled, y_train_scaled)
end_time = time.time()
total_time_rf = end_time - start_time
```

```python
y_pred_rf = rf.predict(X_test_scaled)
```

```python
rmse_rf = mean_squared_error(y_test_scaled, y_pred_rf, squared=False)
rmse_rf
```

```python
print("Tempo total de execução: {:.2f} segundos".format(total_time_rf))
```

### Árvore de Decisão

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
```

```python
param_grid = {
    'max_depth': [2, 4, 6, 8, 10],
    'min_samples_split': [2, 4, 6, 8, 10]
}
```

```python
grid_search = GridSearchCV(dtr, param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train_scaled)
```

```python
print('Melhores hiperparâmetros:', grid_search.best_params_)
print('Melhor pontuação:', grid_search.best_score_)
```

```python
# Treinando o modelo
dtr = DecisionTreeRegressor(max_depth = 10, min_samples_split = 10, random_state=12345)

start_time = time.time()
dtr.fit(X_train_scaled, y_train_scaled)
end_time = time.time()
total_time_dtr = end_time - start_time
```

```python
y_pred_dtr = dtr.predict(X_test_scaled)
```

```python
rmse_dtr = mean_squared_error(y_test_scaled, y_pred_dtr, squared=False)
rmse_dtr
```

```python
print("Tempo total de execução: {:.2f} segundos".format(total_time_dtr))
```

### Modelo Regressão Linear

```python
rl = LinearRegression()

start_time = time.time()
rl.fit(X_train_scaled, y_train_scaled)
end_time = time.time()
total_time_rl = end_time - start_time
```

```python
y_pred_rl = rl.predict(X_test_scaled)
```
